```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('10001.xlsx')
```

```python
val_qT = df[['value', 'qT']]
val_qT.head(10)
```

```python
val_qT.plot()
```

```python
fig, ax = plt.subplots(figsize = (10, 8), constrained_layout = True)
ax.plot(val_qT.iloc[:,1], val_qT.iloc[:,0], '.-')
ax.set_xlabel('Cross Section')
ax.set_ylabel('qT')
```
